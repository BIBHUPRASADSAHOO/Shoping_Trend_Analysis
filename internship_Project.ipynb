```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/sample_data/shopping_trends.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
# Check the null value is present or not on the dataset
df.isnull().sum()
```

```python
#check the duplicate value
df.duplicated().sum()
```

```python
#Dropping the column that are not required for our analysis
df.drop(['Discount Applied','Promo Code Used'],axis=1,inplace=True)
df
```

```python
df.columns
```

```python
# top 10 customer based on purchase amount
top_10_customers=df.groupby('Customer ID')['Purchase Amount (USD)'].sum().nlargest(10)
print(top_10_customers)
```

```python
# Sorting the column
df.sort_values(by='Purchase Amount (USD)',ascending=False)
```

```python
# Find the unique Customer
df['Customer ID'].nunique()
```

```python
# Item purchased in various category
most_purchase_season=df.groupby(['Category','Item Purchased'])['Item Purchased'].count()
most_purchase_season
```

```python
#count the item based on category
Sorting_category=df['Category'].value_counts().sort_values(ascending=False)
# Sorting
sns.barplot(x=Sorting_category.index,y=Sorting_category.values,palette='viridis')
plt.title('Count of Item Based on Category')
plt.xlabel('Category')
plt.show()
```

```python
# Purchase amount among various item categories based on gender
plt.style.use('ggplot')
plt.figure(figsize=(10,6))
sns.boxplot(x='Category',y='Purchase Amount (USD)',hue='Gender',data=df)
plt.title('Purchase Amount by Item Category and Gender',fontsize=20)
plt.xlabel('Item Category',fontsize=20)
plt.ylabel('Purchase Amount (USD)',fontsize=10)
plt.xticks(rotation=45,fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
# Which Payment method is mostly used
best_Payment_method=df['Payment Method'].value_counts().sort_values(ascending=False)
best_Payment_method
```

```python
#plot the most prefered payment method
plt.figure(figsize=(8,6))
plt.style.use('ggplot')
sns.countplot(x='Payment Method',data=df,order=best_Payment_method.index,palette='viridis')
plt.title('Most Prefered Payment Method',fontsize=10)
plt.xlabel('Payment Method',fontsize=10)
plt.ylabel('No of Customer',fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()

```

```python
#Distribution of customer age in the dataset plotiing histogram
plt.figure(figsize=(8,6))
sns.histplot(df['Age'],bins=6,kde=True)
plt.title('Distribution of Customer Age')
plt.xlabel('Customer Age')
plt.ylabel('No of Customer')
plt.show()
```

```python
# Top 10 location by purchase Amount(USD)
top_10_location=df.groupby('Location')['Purchase Amount (USD)'].sum().sort_values(ascending=False).head(10)
top_10_location

# plot
plt.figure(figsize=(20,6))
sns.barplot(x=top_10_location.index,y=top_10_location.values,palette='viridis')
plt.title('Top 10 Location by Purchase Amount(USD)')
plt.xlabel('Location')
plt.ylabel('Purchase Amount(USD)')
plt.show()
```

```python
#portion of subscription status
subscription_status=df['Subscription Status'].value_counts()
subscription_status

#plot
plt.pie(subscription_status,labels=subscription_status.index,autopct='%1.1f%%',startangle=90,colors=sns.color_palette('pastel'))
plt.title('Subscription Status')
plt.show()
```

```python
# Relationshpi between subscription status and purchase
plt.figure(figsize=(15,6))
sns.countplot(x='Frequency of Purchases',hue='Subscription Status',data=df,palette='viridis')
plt.title('Relationship between Subscription Status and Payment Method')
plt.xlabel('Frequency of Purchases')
plt.ylabel('Subscription Status')
plt.show()
#
```
